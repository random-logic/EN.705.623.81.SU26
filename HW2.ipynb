### [25 points] Part A: Implement the Algorithm

Write a dynamic programming solution that computes both:
- The **length** of the longest palindromic subsequence (LPS) in a given string `s`
- The **actual subsequence string** itself

You must:
- Use a **bottom-up (iterative)** dynamic programming approach (not recursion)
- Use a 2D `dp` table where `dp[i][j]` stores the length of the LPS in substring `s[i...j]`
- Include a **traceback routine** to reconstruct one longest palindromic subsequence

Your function should follow this signature:

```python
def longest_palindromic_subsequence(s: str) -> tuple[int, str]:
    pass
```

**A sample input and output is below**

```

s = "bbbab"
dp[0][4] = 4

Traceback path (example):

s[0] = 'b', s[4] = 'b' → match → result = ['b' ... 'b']
    dp[1][3] = 2
s[1] = 'b', s[3] = 'b' → match → result = ['b', 'b' ... 'b', 'b']
    dp[2][2] = 1 (base case)
Final result: "bbbb"
```

````python
def longest_palindromic_sequence(s: str) -> tuple[int, str]:
    # Initialize the dp state
    n = len(s)
    dp: list[list[int]] = [[0] * n for _ in range(n)] # O(n^2)

    # Dynamic programming solution
    # Outer loop runs in O(n) time
    # Total O(n^2) for lines 9-22
    for j in range(0, n):
        for i in range(j, -1, -1): # O(n)
            # Base case
            if i == j: # O(1)
                dp[i][j] = 1 # O(1)
                continue

            # Calculate the longest palindrome subsequence
            if s[i] == s[j]: # O(1)
                # The end characters match
                dp[i][j] = dp[i + 1][j - 1] + 2 # O(1)
            else:
                # The end characters don't match
                dp[i][j] = max(dp[i][j - 1], dp[i + 1][j]) # O(1)

    # Backtracking
    i, j = 0, n - 1 # O(1)
    trace: list[str] = [] # O(1)

    # Condition checking in while loop runs in O(1) time
    # While loop runs in O(n) time
    # Total O(n^2) for lines 31-43
    while i < j:
        if s[i] == s[j]: # O(1)
            trace.append(s[i]) # O(1)
            print(f's[{i}] == s[{j}] -> match -> result = {[*trace, '...', *trace[::-1]]}') # O(n)
            i += 1 # O(1)
            j -= 1 # O(1)
        elif dp[i + 1][j] > dp[i][j - 1]: # O(1)
            # Get the traceback of dp[i + 1][j]
            # If it results in the longest string
            i += 1 # O(1)
        else:
            # Otherwise get the traceback of dp[i][j - 1]
            j -= 1 # O(1)

    if i == j: # O(1)
        # Odd length palindrome
        res = ''.join([*trace, s[i], *trace[::-1]]) # O(n)
        print(f's[{i}] == s[{j}] -> match -> final result = "{res}"') # O(1)
    else:
        # Even length palindrome
        res = ''.join([*trace, *trace[::-1]]) # O(n)
        print(f'final result = "{res}"') # O(1)

    return dp[0][n - 1], res # O(1)

# Test
longest_palindromic_sequence('bbbab')
````

### Part B: Answer
#### Step 1: Defining the State
The dp state, $\text{dp}[i][j]$, will represent the longest palindromic sequence from indices $i$ to $j$.

#### Step 2: Define the Decisions
At any indices $i, j$, we can choose to either add the characters at these indices to our current palindrome or to refuse these characters. We can only add these characters if these do not break the palindromic property, which means the characters have to be equal. If we add them, the length of the palindromic sequence increases by two, and we need to check the longest palindromic sequence of the string from $i + 1$ to $j - 1$. We can also choose to use the longest palindromic sequence by deleting a character, either deleting the character at $i$ or $j$, then checking for the longest palindromic sequence in the substring.

#### Step 3: Define the Recurrence
The recurrence is as follows:
$$
\text{dp}[i][j] =
\begin{cases}
\text{dp}[i+1][j-1] + 2 & \quad s_i = s_j \\
\max\left(
\text{dp}[i+1][j],
\text{dp}[i][j-1]
\right) & \quad s_i \ne s_j
\end{cases}
$$

Where $s_i$, $s_j$ is the characters in the string at $i$-th and $j$-th indices respectively.

Optimal Substructure - This problem has optimal substructure because the longest palindromic subsequence of the current string is going to depend on the longest palindromic subsequence of a substring within the current string.

Overlapping Subproblems - It is possible to compute the same substructure more than once. For example, if we compute `dp[i + 1][j]` then in another round `dp[i + 1][j - 1]`, we would call `dp[i + 1][j - 1]` a second time if in a different round we compute `dp[i][j - 1]` and then `dp[i + 1][j - 1]`.

#### Step 4: Base Cases
The base cases are as follows:
$$
dp[i][j] = 0 \quad \text{for } i > j
$$

$$
dp[i][j] = 1 \quad \text{if } i = j
$$

Whenever $i > j$, there is no valid substring, so $\text{dp}[i][j] = 0$. If $i = j$, $\text{dp}[i][j] = 1$ because a 1 character string is always a palindrome.
