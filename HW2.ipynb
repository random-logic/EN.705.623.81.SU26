# Q1 — Dynamic Programming: Longest Palindromic Subsequence

## 50 points total

### [25 points] Part A: Implement the Algorithm

Write a dynamic programming solution that computes both:
- The **length** of the longest palindromic subsequence (LPS) in a given string `s`
- The **actual subsequence string** itself

You must:
- Use a **bottom-up (iterative)** dynamic programming approach (not recursion)
- Use a 2D `dp` table where `dp[i][j]` stores the length of the LPS in substring `s[i...j]`
- Include a **traceback routine** to reconstruct one longest palindromic subsequence

Your function should follow this signature:

```python
def longest_palindromic_subsequence(s: str) -> tuple[int, str]:
    pass
```

**A sample input and output is below**

```

s = "bbbab"
dp[0][4] = 4

Traceback path (example):

s[0] = 'b', s[4] = 'b' → match → result = ['b' ... 'b']
    dp[1][3] = 2
s[1] = 'b', s[3] = 'b' → match → result = ['b', 'b' ... 'b', 'b']
    dp[2][2] = 1 (base case)
Final result: "bbbb"
```

````python
def longest_palindromic_sequence(s: str) -> tuple[int, str]:
    # In case the string is empty
    if s == "":
        return 0, ""

    # Initialize the dp state
    n = len(s)
    dp: list[list[int]] = [[0] * n for _ in range(n)] # O(n^2)

    # Dynamic programming solution
    # Outer loop runs in O(n) time
    # Total O(n^2) for lines 9-22
    for j in range(0, n):
        for i in range(j, -1, -1): # O(n)
            # Base case
            if i == j: # O(1)
                dp[i][j] = 1 # O(1)
                continue

            # Calculate the longest palindrome subsequence
            if s[i] == s[j]: # O(1)
                # The end characters match
                dp[i][j] = dp[i + 1][j - 1] + 2 # O(1)
            else:
                # The end characters don't match
                dp[i][j] = max(dp[i][j - 1], dp[i + 1][j]) # O(1)

    # Backtracking
    i, j = 0, n - 1 # O(1)
    trace: list[str] = [] # O(1)

    # Condition checking in while loop runs in O(1) time
    # While loop runs in O(n) time, BUT:
    # Total O(n^2) for backtracking on lines 36-57 because of debugging printing
    # Without the debugging printing, backtracking itself just O(n)
    while i < j:
        if s[i] == s[j]: # O(1)
            trace.append(s[i]) # O(1)
            print(f's[{i}] == s[{j}] -> match -> result = {[*trace, '...', *trace[::-1]]}') # O(n)
            i += 1 # O(1)
            j -= 1 # O(1)
        elif dp[i + 1][j] > dp[i][j - 1]: # O(1)
            # Get the traceback of dp[i + 1][j]
            # If it results in the longest string
            i += 1 # O(1)
        else:
            # Otherwise get the traceback of dp[i][j - 1]
            j -= 1 # O(1)

    if i == j: # O(1)
        # Odd length palindrome
        res = ''.join([*trace, s[i], *trace[::-1]]) # O(n)
        print(f's[{i}] == s[{j}] -> match -> final result = "{res}"') # O(1)
    else:
        # Even length palindrome
        res = ''.join([*trace, *trace[::-1]]) # O(n)
        print(f'final result = "{res}"') # O(1)

    return dp[0][n - 1], res # O(1)

# Test
longest_palindromic_sequence('bbbab')
````

### Part B: Answer

#### Step 1: Defining the State

Let $\text{dp}[i][j]$ represent the length of the longest palindromic subsequence in the substring $s[i \dots j]$, where $i$ and $j$ are inclusive indices.

---

#### Step 2: Define the Decisions

At each pair of indices $i$ and $j$, we compare the characters $s[i]$ and $s[j]$.

If $s[i] = s[j]$, then these two characters can be used as the matching outer characters of a palindrome. In that case, the length of the longest palindromic subsequence is:

$$
2 + \text{the longest palindromic subsequence inside } s[i+1 \dots j-1]
$$

So we use:

$$
\text{dp}[i+1][j-1] + 2
$$

If $s[i] \neq s[j]$, then both characters cannot be used together as matching ends of the same palindrome. Therefore, we have two choices:

* Ignore the character at index $i$ and solve the problem for $s[i+1 \dots j]$
* Ignore the character at index $j$ and solve the problem for $s[i \dots j-1]$

We take the better of these two choices.

---

#### Step 3: Define the Recurrence

The recurrence relation is:

$$
\text{dp}[i][j] =
\begin{cases}
\text{dp}[i+1][j-1] + 2 & \text{if } s[i] = s[j] \\
\max(\text{dp}[i+1][j], \text{dp}[i][j-1]) & \text{if } s[i] \ne s[j]
\end{cases}
$$

where $s[i]$ and $s[j]$ are the characters at indices $i$ and $j$.

This recurrence works because the longest palindromic subsequence of a string depends on the longest palindromic subsequences of its smaller substrings. Therefore, the problem has optimal substructure.

The problem also has overlapping subproblems because the same substrings are considered multiple times. For example, both $\text{dp}[i+1][j]$ and $\text{dp}[i][j-1]$ may depend on $\text{dp}[i+1][j-1]$.

---

#### Step 4: Base Cases

For every single-character substring:

$$
\text{dp}[i][i] = 1
$$

where $i$ is any valid index in the string. This is because any one-character string is always a palindrome of length $1$.

For an invalid substring, where $i > j$:

$$
\text{dp}[i][j] = 0
$$

This represents an empty substring, so the longest palindromic subsequence has length $0$.

---

#### Step 5: Time Complexity

Let

$$
n = \text{len}(s)
$$

The table $\text{dp}$ has $n$ rows and $n$ columns, so it contains:

$$
n \times n = n^2
$$

entries.

The algorithm fills in the values of $\text{dp}[i][j]$ for all valid pairs of indices $i$ and $j$ where $0 \leq i \leq j < n$. There are $O(n^2)$ such pairs.

For each entry $\text{dp}[i][j]$, the work is constant:

* Compare $s[i]$ and $s[j]$
* Either compute $\text{dp}[i+1][j-1] + 2$
* Or compute $\max(\text{dp}[i+1][j], \text{dp}[i][j-1])$

Each of these operations takes $O(1)$ time.

Therefore, filling the dynamic programming table takes:

$$
O(n^2)
$$

After the table is filled, the traceback starts at $\text{dp}[0][n-1]$ and moves through the table. At each step, either $i$ increases, $j$ decreases, or both happen. Since $i$ can increase at most $n$ times and $j$ can decrease at most $n$ times, the traceback takes:

$$
O(n)
$$

Thus, the total time complexity is:

$$
O(n^2) + O(n) = O(n^2)
$$

The $O(n^2)$ table-filling step dominates the $O(n)$ traceback step.

---

#### Step 6: Space Complexity

The algorithm stores the dynamic programming table $\text{dp}$.

Since $\text{dp}$ has $n$ rows and $n$ columns, the table requires:

$$
n \times n = n^2
$$

space.

Therefore, the space complexity for the table is:

$$
O(n^2)
$$

During traceback, the algorithm also stores characters from the first half of the palindrome in a list, such as `trace`. In the worst case, this list can store up to $O(n)$ characters.

So the extra traceback storage is:

$$
O(n)
$$

Combining these:

$$
O(n^2) + O(n) = O(n^2)
$$

Therefore, the overall space complexity is:

$$
O(n^2)
$$

because the $\text{dp}$ table dominates the memory usage.

---

#### Step 7: Traceback Explanation

After filling the $\text{dp}$ table, the length of the longest palindromic subsequence for the entire string $s[0 \dots n-1]$ is stored in:

$$
\text{dp}[0][n-1]
$$

Therefore, the traceback starts at the top-right cell of the table:

$$
i = 0, \quad j = n - 1
$$

This corresponds to considering the full string $s[0 \dots n-1]$.

The goal of traceback is to reconstruct one actual longest palindromic subsequence, not just its length.

At each step, we compare $s[i]$ and $s[j]$.

If:

$$
s[i] = s[j]
$$

then the characters at indices $i$ and $j$ can be used as matching outer characters of the palindrome. In this case, we add $s[i]$ to a list called `trace`, which stores the first half of the palindrome. Then we move inward:

$$
i \leftarrow i + 1
$$

$$
j \leftarrow j - 1
$$

If:

$$
s[i] \ne s[j]
$$

then the two characters cannot be used together as matching outer characters. In this case, we look at the neighboring table entries:

$$
\text{dp}[i+1][j]
$$

and

$$
\text{dp}[i][j-1]
$$

These represent the best answers after ignoring either $s[i]$ or $s[j]$.

If:

$$
\text{dp}[i+1][j] > \text{dp}[i][j-1]
$$

then the longer subsequence comes from ignoring $s[i]$, so we move down in the table:

$$
i \leftarrow i + 1
$$

Otherwise, the longer subsequence comes from ignoring $s[j]$, or both choices are tied. In that case, we move left in the table:

$$
j \leftarrow j - 1
$$

This process continues until:

$$
i \geq j
$$

The list `trace` stores the first half of the palindrome. Since a palindrome is symmetric, the second half can be created by reversing `trace`.

If the traceback ends with:

$$
i = j
$$

then there is one middle character, so the palindrome has odd length. The final palindrome is constructed as:

```trace + trace[::-1]```

For example, for the string "bbbab", one longest palindromic sequence is "bbbb". The traceback finds the matching characters, stores the first half, and then mirrors it to produce the full palindrome.

## Minimax algorithm pseudocode

```text
function Min-max-Search(game, state) returns an action
    player ← game.To-Move(state)
    value, move ← Max-Value(game, state)
    return move

function Max-Value(game, state) returns a (utility, move) pair
    if game.Is-Terminal(state) then
        return game.Utility(state, player), null
    v ← −∞
    for each a in game.Actions(state) do
        v2, a2 ← Min-Value(game, game.Result(state, a))
        if v2 > v then
            v, move ← v2, a
    return v, move

function Min-Value(game, state) returns a (utility, move) pair
    if game.Is-Terminal(state) then
        return game.Utility(state, player), null
    v ← +∞
    for each a in game.Actions(state) do
        v2, a2 ← Max-Value(game, game.Result(state, a))
        if v2 < v then
            v, move ← v2, a
    return v, move

## Alpha–Beta search algorithm pseudocode

```text
function Alpha-Beta-Decision(game, state) returns an action
    player ← game.To-Move(state)
    value, move ← Max-Value(game, state, −∞, +∞)
    return move

function Max-Value(game, state, α, β) returns a (utility, move) pair
    if game.Is-Terminal(state) then
        return game.Utility(state, player), null
    v ← −∞
    for each a in game.Actions(state) do
        v2, a2 ← Min-Value(game, game.Result(state, a), α, β)
        if v2 > v then
            v, move ← v2, a
        α ← Max(α, v)
        if v ≥ β then
            return v, move
    return v, move

function Min-Value(game, state, α, β) returns a (utility, move) pair
    if game.Is-Terminal(state) then return game.Utility(state, player), null
    v ← +∞
    for each a in game.Actions(state) do
        v2, a2 ← Max-Value(game, game.Result(state, a), α, β)
        if v2 < v then
            v, move ← v2, a
        β ← Min(β, v)
        if v ≤ α then
            return v, move
    return v, move

````python
from graphviz import Digraph
import math

# ------------------------------------------------------------
# Board indexing used in the problem:
#
# 0 | 3 | 6
# 1 | 4 | 7
# 2 | 5 | 8
# ------------------------------------------------------------

WIN_LINES = [
    (0, 3, 6),  # top row
    (1, 4, 7),  # middle row
    (2, 5, 8),  # bottom row
    (0, 1, 2),  # left column
    (3, 4, 5),  # middle column
    (6, 7, 8),  # right column
    (0, 4, 8),  # diagonal
    (6, 4, 2),  # diagonal
]

def board_to_string(board):
    """
    Converts board dictionary/list into the visual layout:
    0 | 3 | 6
    1 | 4 | 7
    2 | 5 | 8
    """
    def cell(i):
        return board[i] if board[i] != "." else "."

    return (
        f"{cell(0)} | {cell(3)} | {cell(6)}\n"
        f"{cell(1)} | {cell(4)} | {cell(7)}\n"
        f"{cell(2)} | {cell(5)} | {cell(8)}"
    )

def actions(board):
    return [i for i in range(9) if board[i] == "."]

def winner(board):
    for a, b, c in WIN_LINES:
        if board[a] != "." and board[a] == board[b] == board[c]:
            return board[a]
    return None

def is_terminal(board):
    return winner(board) is not None or len(actions(board)) == 0

def utility(board):
    """
    Utility is from X's perspective.
    X win = +1
    draw = 0
    O win = -1
    """
    w = winner(board)
    if w == "X":
        return 1
    elif w == "O":
        return -1
    else:
        return 0

def result(board, move, player):
    new_board = board.copy()
    new_board[move] = player
    return new_board

def next_player(player):
    return "O" if player == "X" else "X"

def node_color(player, terminal=False, value=None, pruned=False):
    if pruned:
        return "#dddddd"
    if terminal:
        if value == 1:
            return "#c8f7c5"   # green for X win
        elif value == -1:
            return "#f7c5c5"   # red for O win
        else:
            return "#fff4b8"   # yellow for draw
    if player == "X":
        return "#cce5ff"       # blue-ish for Max/X
    else:
        return "#ffd6cc"       # orange-ish for Min/O

# ------------------------------------------------------------
# Minimax tree construction
# ------------------------------------------------------------

node_counter = 0

def new_node_id():
    global node_counter
    node_counter += 1
    return f"N{node_counter}"

def minimax_build(dot, board, player_to_move, depth=0, incoming_move=None):
    """
    Builds the full minimax tree.
    Returns:
        node_id, minimax_value, best_move
    """
    node_id = new_node_id()

    terminal = is_terminal(board)

    if terminal:
        val = utility(board)
        label = (
            f"Terminal\n"
            f"{board_to_string(board)}\n\n"
            f"winner = {winner(board) or 'Draw'}\n"
            f"utility = {val}"
        )
        dot.node(
            node_id,
            label=label,
            shape="box",
            style="filled",
            fillcolor=node_color(player_to_move, terminal=True, value=val),
            fontname="Courier"
        )
        return node_id, val, None

    acts = actions(board)

    if player_to_move == "X":
        # Max node
        v = -math.inf
        best_move = None
        node_type = "Max-Value"
    else:
        # Min node
        v = math.inf
        best_move = None
        node_type = "Min-Value"

    # Create placeholder node first
    label_before = (
        f"{node_type}\n"
        f"player = {player_to_move}\n"
        f"{board_to_string(board)}\n\n"
        f"Actions = {acts}\n"
        f"v starts = {'-∞' if player_to_move == 'X' else '+∞'}"
    )

    dot.node(
        node_id,
        label=label_before,
        shape="box",
        style="filled",
        fillcolor=node_color(player_to_move),
        fontname="Courier"
    )

    for move in acts:
        child_board = result(board, move, player_to_move)
        child_id, child_val, _ = minimax_build(
            dot,
            child_board,
            next_player(player_to_move),
            depth + 1,
            incoming_move=move
        )

        dot.edge(
            node_id,
            child_id,
            label=f"{player_to_move}:{move}\nvalue={child_val}",
            fontname="Courier"
        )

        if player_to_move == "X":
            if child_val > v:
                v = child_val
                best_move = move
        else:
            if child_val < v:
                v = child_val
                best_move = move

    # Update node with final minimax result
    final_label = (
        f"{node_type}\n"
        f"player = {player_to_move}\n"
        f"{board_to_string(board)}\n\n"
        f"Actions = {acts}\n"
        f"final v = {v}\n"
        f"best move = {best_move}"
    )

    dot.node(
        node_id,
        label=final_label,
        shape="box",
        style="filled",
        fillcolor=node_color(player_to_move),
        fontname="Courier"
    )

    return node_id, v, best_move

# ------------------------------------------------------------
# Alpha-beta tree construction
# ------------------------------------------------------------

def fmt(x):
    if x == math.inf:
        return "+∞"
    if x == -math.inf:
        return "-∞"
    return str(x)

def alphabeta_build(dot, board, player_to_move, alpha, beta, depth=0):
    """
    Builds alpha-beta tree.
    Returns:
        node_id, value, best_move
    """
    node_id = new_node_id()
    terminal = is_terminal(board)

    if terminal:
        val = utility(board)
        label = (
            f"Terminal\n"
            f"{board_to_string(board)}\n\n"
            f"winner = {winner(board) or 'Draw'}\n"
            f"utility = {val}\n"
            f"α={fmt(alpha)}, β={fmt(beta)}"
        )
        dot.node(
            node_id,
            label=label,
            shape="box",
            style="filled",
            fillcolor=node_color(player_to_move, terminal=True, value=val),
            fontname="Courier"
        )
        return node_id, val, None

    acts = actions(board)

    if player_to_move == "X":
        node_type = "Max-Value"
        v = -math.inf
    else:
        node_type = "Min-Value"
        v = math.inf

    best_move = None
    start_alpha = alpha
    start_beta = beta

    label_before = (
        f"{node_type}\n"
        f"player = {player_to_move}\n"
        f"{board_to_string(board)}\n\n"
        f"Actions = {acts}\n"
        f"v starts = {fmt(v)}\n"
        f"α={fmt(alpha)}, β={fmt(beta)}"
    )

    dot.node(
        node_id,
        label=label_before,
        shape="box",
        style="filled",
        fillcolor=node_color(player_to_move),
        fontname="Courier"
    )

    for idx, move in enumerate(acts):
        child_board = result(board, move, player_to_move)
        child_id, child_val, _ = alphabeta_build(
            dot,
            child_board,
            next_player(player_to_move),
            alpha,
            beta,
            depth + 1
        )

        dot.edge(
            node_id,
            child_id,
            label=f"{player_to_move}:{move}\nvalue={child_val}",
            fontname="Courier"
        )

        if player_to_move == "X":
            if child_val > v:
                v = child_val
                best_move = move
            alpha = max(alpha, v)

            # Beta cutoff
            if v >= beta:
                # Show remaining unexplored actions, if any
                remaining = acts[idx + 1:]
                for r in remaining:
                    pruned_id = new_node_id()
                    pruned_label = (
                        f"PRUNED\n"
                        f"move {player_to_move}:{r}\n"
                        f"because v={v} >= β={fmt(beta)}"
                    )
                    dot.node(
                        pruned_id,
                        label=pruned_label,
                        shape="box",
                        style="filled,dashed",
                        fillcolor=node_color(player_to_move, pruned=True),
                        fontname="Courier"
                    )
                    dot.edge(
                        node_id,
                        pruned_id,
                        label=f"{player_to_move}:{r}",
                        style="dashed",
                        fontname="Courier"
                    )
                break

        else:
            if child_val < v:
                v = child_val
                best_move = move
            beta = min(beta, v)

            # Alpha cutoff
            if v <= alpha:
                # Show remaining unexplored actions, if any
                remaining = acts[idx + 1:]
                for r in remaining:
                    pruned_id = new_node_id()
                    pruned_label = (
                        f"PRUNED\n"
                        f"move {player_to_move}:{r}\n"
                        f"because v={v} <= α={fmt(alpha)}"
                    )
                    dot.node(
                        pruned_id,
                        label=pruned_label,
                        shape="box",
                        style="filled,dashed",
                        fillcolor=node_color(player_to_move, pruned=True),
                        fontname="Courier"
                    )
                    dot.edge(
                        node_id,
                        pruned_id,
                        label=f"{player_to_move}:{r}",
                        style="dashed",
                        fontname="Courier"
                    )
                break

    final_label = (
        f"{node_type}\n"
        f"player = {player_to_move}\n"
        f"{board_to_string(board)}\n\n"
        f"Actions = {acts}\n"
        f"start α={fmt(start_alpha)}, start β={fmt(start_beta)}\n"
        f"final α={fmt(alpha)}, final β={fmt(beta)}\n"
        f"final v = {v}\n"
        f"best move = {best_move}"
    )

    dot.node(
        node_id,
        label=final_label,
        shape="box",
        style="filled",
        fillcolor=node_color(player_to_move),
        fontname="Courier"
    )

    return node_id, v, best_move

# ------------------------------------------------------------
# Build initial state after X plays position 5
# ------------------------------------------------------------

# Original Figure 1:
# . | . | O
# O | O | X
# X | . | .
#
# Positions:
# 0 | 3 | 6
# 1 | 4 | 7
# 2 | 5 | 8

initial_board = ["."] * 9
initial_board[6] = "O"
initial_board[1] = "O"
initial_board[4] = "O"
initial_board[7] = "X"
initial_board[2] = "X"

# X takes the next move at position 5
initial_board[5] = "X"

# Now O is to move
player_to_move = "O"

print("Initial board after X moves at 5:")
print(board_to_string(initial_board))
print("Actions:", actions(initial_board))

# ------------------------------------------------------------
# Generate minimax graph
# ------------------------------------------------------------

node_counter = 0
minimax_dot = Digraph("Minimax_Tree", format="png")
minimax_dot.attr(rankdir="TB")
minimax_dot.attr("node", fontsize="10")
minimax_dot.attr("edge", fontsize="9")

root_id, value, best_move = minimax_build(
    minimax_dot,
    initial_board,
    player_to_move
)

print("\nMinimax result:")
print("value =", value)
print("best move =", best_move)

minimax_dot.render("minimax_tree", cleanup=True)

# ------------------------------------------------------------
# Generate alpha-beta graph
# ------------------------------------------------------------

node_counter = 0
ab_dot = Digraph("AlphaBeta_Tree", format="png")
ab_dot.attr(rankdir="TB")
ab_dot.attr("node", fontsize="10")
ab_dot.attr("edge", fontsize="9")

root_id, value, best_move = alphabeta_build(
    ab_dot,
    initial_board,
    player_to_move,
    alpha=-math.inf,
    beta=math.inf
)

print("\nAlpha-beta result:")
print("value =", value)
print("best move =", best_move)

ab_dot.render("alphabeta_tree", cleanup=True)

print("\nGenerated files:")
print("minimax_tree.png")
print("alphabeta_tree.png")
````
