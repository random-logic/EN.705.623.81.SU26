# 2
### (a) Variables, Domains, and Constraints
#### Variables
Let:
$$x_{i,j} =
\begin{cases}
1 & \text{if there is a queen in row } i \text{ and column } j \\
0 & \text{otherwise}
\end{cases}$$
Where $i,j \in [1, 3]$ are the indices of rows and columns.

There are 9 binary variables:
$$x_{1,1}, x_{1,2}, x_{1,3}, x_{2,1}, x_{2,2}, x_{2,3}, x_{3,1}, x_{3,2}, x_{3,3}$$
Each variable represents whether a queen is placed in a particular square of the $3 \times 3$ board.

#### Domain
Each variable has domain:
$$x_{ij} \in \{0,1\}$$

Where:
* $1$ means a queen is placed in that cell.
* $0$ means no queen is placed in that cell.

#### Constraints
The constraints are:
* Exactly one queen per row.
* Exactly one queen per column.
* At most one queen per diagonal.

### (b) Row Constraints
For rows, this can be rewritten as:
$$\left(\forall i \in [1, 3]\right)\left(\sum_{j = 1}^3 x_{i, j} = 1 \right)$$

### (c) Column Constraints
And for columns:
$$\left(\forall i \in [1, 3]\right)\left(\sum_{j = 1}^3 x_{i, j} = 1 \right)$$

### (d) One Diagonal Constraint
Rewrite the diagonal constraint as:
$$\sum_{j = 1}^3 x_{j, j} \leq 1$$

This is an inequality constraint rather than an equality constraint because the diagonal may contain zero or one queen, but not more than one.

### (e) One Row and One Column Constraint as Squared Violation Terms
For row $1$, the squared violation term is:
$$\left(x_{11} + x_{12} + x_{13} - 1\right)^2$$

And for column $1$:
$$\left(x_{11} + x_{21} + x_{31} - 1\right)^2$$

If the constraints are satisfied, these terms equal 0. If the constraints are violated, the terms are positive.

### (f) Construct a Symbolic Total Violation Function $E$
A total violation function combines the violations of all constraints into one expression. For the row and column constraints:
$$E_{\text{rows}} =
\sum_{i = 1}^3 \left(\left(\sum_{j = 1}^3 x_{i, j}\right)-1\right)^2$$
$$E_{\text{cols}} =
\sum_{i = 1}^3 \left(\left(\sum_{j = 1}^3 x_{j, i}\right)-1\right)^2$$

For diagonal constraints, we need to penalize cases where more than one queen appears on a diagonal. One way to write this is:
$$E_{\text{diags}} =
\sum_{d} \max(0, S_d - 1)^2$$
Where $S_d$ is the sum of the variables on diagonal $d$.

### (g) Why $E = 0$ Corresponds to a Feasible Solution
The total violation function $E$ is a sum of squared violation terms.
Each term is always nonnegative. Therefore, $E = 0$, only if every individual violation term is also $0$.

This means all the constraints are satisfied by definition of our violation terms. That is:
* Every row has exactly one queen.
* Every column has exactly one queen.
* No diagonal has more than one queen.

However, for the $3 \times 3$ N-Queens problem, no feasible solution actually exists. Therefore, in this specific case, the best possible value of $E$ would be greater than 0.

### (h) How This Differs From Solving the CSP Via Backtracking
Backtracking solves the CSP by assigning values to variables one at a time and checking constraints after each partial assignment. For example, it might place a queen in Row 1, then try Row 2, then Row 3, and backtrack when it finds a conflict. The algebraic formulation is different because it turns the problem into minimizing a function $E$. Instead of explicitly searching through assignments and undoing choices, we evaluate how badly an assignment violates the constraints and try to reduce that violation via optimization.

### (i) Optimization Explanation
This formulation is closest to optimization. The original CSP is usually described logically as "find an assignment that satisfies all constraints". But after reformulating it as a total violation function $E$, the goal becomes minimize $E$, where a perfect solution has $E = 0$. So the problem becomes minimizing constraint violations rather than directly performing logical inference or traditional backtracking search.

# 3 Simulate Backtracking with Forward Checking

Variables: $A, B, C$

Domains: $\{R, G\}$ for each variable

Constraints: $A \neq B$, $B \neq C$

Variable order: $A \rightarrow B \rightarrow C$

Value order: $R$ before $G$


### Initial State

Current assignment:

${}$

Domains:

| Variable | Domain     |
|----------|------------|
| A        | $\{R, G\}$ |
| B        | $\{R, G\}$ |
| C        | $\{R, G\}$ |

Action: Continue

### Step 1: Try $A = R$

Current assignment:

$\{A = R\}$

Forward checking using constraint $A \neq B$:

Since $A = R$, remove $R$ from $B$'s domain.

Updated domains:

| Variable | Domain     |
|----------|------------|
| A        | $\{R\}$    |
| B        | $\{G\}$    |
| C        | $\{R, G\}$ |

Action: Continue

### Step 2: Try $B = G$

Current assignment:

$\{A = R, B = G\}$

Forward checking using constraint $B \neq C$:

Since $B = G$, remove $G$ from $C$'s domain.

Updated domains:

| Variable | Domain  |
|----------|---------|
| A        | $\{R\}$ |
| B        | $\{G\}$ |
| C        | $\{R\}$ |

Action: Continue

### Step 3: Try $C = R$

Current assignment:

$\{A = R, B = G, C = R\}$

There are no unassigned neighboring variables left to update.

Updated domains:

| Variable | Domain  |
|----------|---------|
| A        | $\{R\}$ |
| B        | $\{G\}$ |
| C        | $\{R\}$ |

Action: Solution found

---

### Final Solution

$\{A = R, B = G, C = R\}$
