### Merge Sort
Answer: NOT a DP problem

#### Explanation
Merge sort is a divide-and-conquer algorithm. It divides the array into independent left and right subarrays, recursively sorts each half, and then combines the results by merging. It has optimal substructure in the sense that sorting the whole array depends on correctly sorting its smaller parts. However, the subproblems do not overlap because each subarray is processed only once. Therefore, memoization would not provide a benefit, so merge sort is not considered a dynamic programming problem.

### Towers of Hanoi
Answer: NOT a DP problem

#### Explanation
The Towers of Hanoi is a pure recursion problem. At any $n$-th block, the solution involves the following:

1. Moving disks $1$ to $n - 1$ to the auxiliary pillar. This is done recursively by calling the same function on the $n - 1$ disks, but instead using the auxiliary pillar as the target and the target pillar as the auxiliary.
2. Moving the $n$-th disk to the target pillar.
3. Moving disks $1$ to $n - 1$ to the target pillar. This is done recursively by calling the same function on the $n - 1$ disks, but instead using the auxiliary pillar as the source and the source pillar as the auxiliary.

This problem has optimal substructure if it is viewed as minimizing the number of moves, since the optimal solution for $n$ disks depends on optimal solutions for moving $n - 1$ disks. The recursive calls are structurally similar, but because they involve different source, target, and auxiliary pillars, they are not overlapping subproblems. Therefore, memoization would not provide a benefit, so the Towers of Hanoi problem is not a dynamic programming problem.

### Minimum Coin Change
Answer: DP problem

#### Explanation
This problem has optimal substructure because if a coin of value $c$ is chosen, then the remaining problem is to find the minimum number of coins needed to make $\text{target} - c$. The recurrence is:

$$
\text{dp}[\text{target}] = 1 + \min_{c \in \text{coins}, c \leq \text{target}} \text{dp}[\text{target} - c]
$$

Where:
* $\text{dp}[\text{target}]$ represents the minimum number of coins needed to reach the target amount.
* $1$ represents the coin that was just chosen.
* $\min_{c \in \text{coins}, c \leq \text{target}} \text{dp}[\text{target} - c]$ represents the minimum number of coins needed to make the remaining amount after choosing one valid coin of value $c$, considering all possible coin choices.

There are overlapping subproblems because the same smaller amounts may be computed multiple times in a naive recursive solution. For example, when solving $\text{dp}[15]$, the subproblem $\text{dp}[10]$ can be reached by choosing a coin of value $5$ or by choosing five coins of value $1$ through a sequence of recursive calls. Therefore, memoization or tabulation can improve efficiency, so dynamic programming is helpful.

### DP State
The DP State represents the number of ways to climb to the top given $n$ steps.

### Recurrence Relation
To reach step $n$, the final move must either be a one-step move from step $n - 1$ or a two-step move from step $n - 2$. So the recurrence relation for this problem is:

$$\text{dp}[n] = \text{dp}[n - 1] + \text{dp}[n - 2]$$

Where:
* $\text{dp}[n]$ represents the number of ways to climb $n \geq 2$ steps.
* $\text{dp}[n - 1]$ represents the number of ways to reach step $n - 1$.
* $\text{dp}[n - 2]$ represents the number of ways to reach step $n - 2$.

### Base Cases
There is exactly one way to climb $0$ steps - standing still, because there is nothing to climb. Similarly, there is exactly one way to climb $1$ step - take one step. So:

$$\text{dp}[0] = 1\text{, dp}[1] = 1$$

### Optimal Substructure
This problem has optimal substructure, because any valid way to reach step $n$ must come from either step $n - 1$ using a 1-step move or step $n - 2$ using a 2-step move.

### Overlapping Subproblems
There are overlapping subproblems because the same smaller values of $\text{dp}[k]$ for $k < n$ may be computed multiple times in a naive recursive solution. For example, to compute $\text{dp}[5]$, we need $\text{dp}[4]$ and $\text{dp}[3]$. Then computing $\text{dp}[4]$ also requires $\text{dp}[3]$, so $\text{dp}[3]$ is computed more than once without memoization.

### Comparing Top-Down vs Bottom-Up

```python
# Top-Down (Memoized)
# Total runtime: O(n)
def climb_td(n, memo):
    if n in memo: # O(1)
        return memo[n] # O(1)
    memo[n] = climb_td(n-1, memo) + climb_td(n-2, memo) # T(n - 1) + T(n - 2)
    return memo[n] # O(1)

# Bottom-Up (Tabulated)
# Total runtime: O(n)
def climb_bu(n):
    dp = [0] * (n+1) # O(1)
    dp[0], dp[1] = 1, 1 # O(1)
    for i in range(2, n+1): # O(n)
        dp[i] = dp[i-1] + dp[i-2] # O(1)
    return dp[n] # O(1)
```

#### Similarities
Both the top-down and bottom-up solutions store solutions to smaller subproblems, so those values do not need to be recomputed repeatedly. As a result, both run in $O(n)$ time.

#### Differences
The top-down memoized approach starts from the original problem, $\text{dp}[n]$, and recursively breaks it down into smaller subproblems like $\text{dp}[n-1]$, $\text{dp}[n-2]$, etc. It computes subproblems only when they are needed. On the other hand, the bottom-up tabulated approach starts from the base cases, $\text{dp}[0]$ and $\text{dp}[1]$, and iteratively builds up to $\text{dp}[n]$.

#### Memory Usage
Both approaches in the **given code** use $O(n)$ memory. However, the top-down memoized approach usually uses more memory in practice because it stores both the memo table and the recursive call stack, while the bottom-up tabulated approach only stores the DP array.

Also, the bottom-up approach can be optimized because the recurrence only depends on the previous two values, $\text{dp}[n-1]$ and $\text{dp}[n-2]$. Instead of storing the entire DP array, we can store only the last two values, reducing the memory usage to $O(1)$.

#### Large $n$
For large $n$, I would choose the bottom-up approach optimized to use $O(1)$ memory. Although the time complexity remains $O(n)$, storing only the last two values significantly reduces memory usage compared to storing the full DP array. It also avoids recursion depth limitations and the extra overhead of recursive function calls.

### Bellman Equation for Distance of Node $B$
The Bellman Equation for the distance of node $B$ is:

$$\text{dist}[B] = \min\left(\text{dist}[B]\text{, }\min_{(v, B) \in E}\left(\text{dist}[v] + w(v, B)\right)\right)$$

Where:

* $\text{dist}[u]$ is the distance to any node $u$.
* $v$ is any node in the graph that has a directed edge to node $B$.
* $E$ is the set of directed edges in the graph.
* $w(v, B)$ is the weight of the edge from any $v$ to node $B$.

Since the only incoming edges to node $B$ are from nodes $A$ and $C$, the updated equation is:

$$\text{dist}[B] = \min(\text{dist}[B], \text{dist}[A] + 4, \text{dist}[C] - 1)$$

### What the Equation is Computing
The equation computes the shortest known distance from the source node to the node being updated. It checks whether reaching that node through one of its incoming neighbors gives a shorter path than the current known distance.

If $A$ is the source node and $B$ is the node being updated, it compares:

* The current known distance to $B$.
* The current known distance to $A$, plus the edge cost from $A$ to $B$.
* The current known distance to $C$, plus the edge cost from $C$ to $B$.

Then it keeps the shortest distance possible to $B$.

### One Bellman Pass Over All Edges

```python
# Given edges
edges = [
    ('A', 'B', 4),
    ('A', 'C', 2),
    ('C', 'B', -1),
    ('B', 'D', 2),
    ('C', 'D', 5)
]

# Initialize the dist, assuming A is the source node
import math
dist = {
    'A': 0,
    'B': math.inf,
    'C': math.inf,
    'D': math.inf,
}

# Update dist using Bellman Equation
def bellman_update(dist, edges):
    for u, v, w in edges:
        if dist[u] + w < dist[v]:
            new_dist = dist[u] + w
            print(f"{u} -> {v}: Distance updated from {dist[v]} to {new_dist}")
            dist[v] = new_dist
    return dist

# Output the initial distnaces
print(f"Initial distances: {dist}")

# Do one bellman update
bellman_update(dist, edges)

# Output the new known distances
print(f"Final distances: {dist}")
```
