### Merge Sort
Answer: NOT a DP problem

#### Explanation
Merge sort is a divide-and-conquer algorithm. It divides the array into independent left and right subarrays, recursively sorts each half, and then combines the results by merging. It has optimal substructure in the sense that sorting the whole array depends on correctly sorting its smaller parts. However, the subproblems do not overlap because each subarray is processed only once. Therefore, memoization would not provide a benefit, so merge sort is not considered a dynamic programming problem.

### Towers of Hanoi
Answer: NOT a DP problem

#### Explanation
The Towers of Hanoi is a pure recursion problem. At any $n$-th block, the solution involves the following:

1. Moving disks $1$ to $n - 1$ to the auxiliary pillar. This is done recursively by calling the same function on the $n - 1$ disks, but instead using the auxiliary pillar as the target and the target pillar as the auxiliary.
2. Moving the $n$-th disk to the target pillar.
3. Moving disks $1$ to $n - 1$ to the target pillar. This is done recursively by calling the same function on the $n - 1$ disks, but instead using the auxiliary pillar as the source and the source pillar as the auxiliary.

This problem has optimal substructure if it is viewed as minimizing the number of moves, since the optimal solution for $n$ disks depends on optimal solutions for moving $n - 1$ disks. The recursive calls are structurally similar, but because they involve different source, target, and auxiliary pillars, they are not overlapping subproblems. Therefore, memoization would not provide a benefit, so the Towers of Hanoi problem is not a dynamic programming problem.

### Minimum Coin Change
Answer: DP problem

#### Explanation
This problem has optimal substructure because if a coin of value $c$ is chosen, then the remaining problem is to find the minimum number of coins needed to make $\text{target} - c$. The recurrence is:

$$
\text{dp}[\text{target}] = 1 + \min_{c \in \text{coins}, c \leq \text{target}} \text{dp}[\text{target} - c]
$$

Where:
* $\text{dp}[\text{target}]$ represents the minimum number of coins needed to reach the target amount.
* $1$ represents the coin that was just chosen.
* $\min_{c \in \text{coins}, c \leq \text{target}} \text{dp}[\text{target} - c]$ represents the minimum number of coins needed to make the remaining amount after choosing one valid coin of value $c$, considering all possible coin choices.

There are overlapping subproblems because the same smaller amounts may be computed multiple times in a naive recursive solution. For example, when solving $\text{dp}[15]$, the subproblem $\text{dp}[10]$ can be reached by choosing a coin of value $5$ or by choosing five coins of value $1$ through a sequence of recursive calls. Therefore, memoization or tabulation can improve efficiency, so dynamic programming is helpful.
