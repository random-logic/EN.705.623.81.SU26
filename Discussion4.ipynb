### 1. When Optimal Is Not an Option (Lecture 3D/4A)
Scenario: GPS navigation with live traffic

#### (a) Objective: What is this system trying to optimize?
The system is trying to find the fastest route to the destination. It may also consider factors like distance, traffic congestion, and road closures, but the main goal is usually to minimize total travel time.

#### (b) Breakdown: Why is computing the globally optimal solution impractical here?
Computing the globally optimal route is impractical because the road network is large and traffic conditions change constantly. If the system takes too long to calculate the perfect route, the traffic situation may already be different by the time the route is returned. Users also expect directions within seconds, so waiting minutes for a slightly better route would make the system less useful and negate time savings.

#### (c) Constraint: Which matters most - Time, memory, environment changes, uncertainty?
Time matters most because the GPS must respond quickly enough to be useful for real-time navigation. Users generally prefer a good route within seconds rather than waiting longer for a route that may only be slightly better. Environment changes are also important because live traffic, accidents, and road closures can alter the best route while the system is calculating. However, these changes usually only require rerouting when they significantly affect the commute. Overall, time remains the most important constraint because it has the greatest impact on usability and the system must make decisions immediately.

#### (d) Decision Shift: If the system must act now, what guarantee must be given up, and what is gained instead?
The system gives up the guarantee of finding the globally optimal route. In exchange, it gains speed and responsiveness, allowing it to provide a good route quickly. This tradeoff is useful in real-world navigation because a fast, mostly accurate route is often better than a perfect route that takes too long to compute.

### 2. The Cost of Greedy Decisions (Lecture 4B)
Algorithm: Greedy Best-First Search

#### (a) Dominant Operation: What operation dominates the runtime of this algorithm?
The dominant operation is selecting and expanding the node with the best heuristic value. The algorithm repeatedly evaluates possible next nodes and chooses the one that appears closest to the goal.

#### (b) Key Complexity Drivers: Identify two factors that most affect runtime: Branching factor, depth/number of iterations, beam width, heuristic evaluation cost.
Two major factors that affect runtime are the branching factor and the heuristic evaluation cost. At each node, a larger branching factor creates more possible nodes to consider and a more expensive heuristic takes longer to calculate.

#### (c) Why It Scales: Explain why this algorithm typically runs faster than exhaustive or exact search.
Greedy Best-First Search usually runs faster than exhaustive search because it focuses only on the most promising paths instead of exploring all possibilities. By using a heuristic, it can ignore large parts of the search space.

#### (d) What It Sacrifices: What guarantee is lost because of this greedy design?
It sacrifices the guarantee of finding the optimal solution. Since it only follows what looks best at the moment, it can be misled by a poor heuristic and choose a longer or worse path.

### 3. Tracing and Stress-Testing A* (Lecture 4C)

#### (a) Fill in the table for the first four node expansions.

| Expanded Node | $g(n)$ | $h(n)$ | $f(n)$ |
|---------------|-------:|-------:|-------:|
| $S$           |    $0$ |    $6$ |    $6$ |
| $A$           |    $2$ |    $4$ |    $6$ |
| $C$           |    $5$ |    $2$ |    $7$ |
| $G$           |    $7$ |    $0$ |    $7$ |

#### (b) Which node is expanded earlier by A* that would not necessarily be expanded early by Greedy Best-First Search? Why?
None. In this scenario, A* and Greedy Best-First Search expand the same nodes early. Greedy Best-First Search selects the node $n$ in the open list with the lowest heuristic value $h(n)$, without considering the path cost $g(n)$ from the starting node $S$ to $n$. This differs from A*, which selects nodes by minimizing $g(n) + h(n)$. However, in this graph, the heuristic values coincidentally happens to guide Greedy Best-First Search along the same route that A* chooses. Therefore, no node is expanded earlier by A*. This may not happen in every scenario, but it does in this one.

#### (c) Now suppose the heuristic for node D is changed from $h(D) = 3$ to $h(D) = 7$
Q: Is the heuristic still admissible? Why or why not?

A: No, because now $h(D) = 7$ overestimates the true remaining optimal cost of $h^*(D) = 5$. An admissible heuristic must satisfy the following constraint for any node $n$:

$$h(n) \leq h^*(n)$$

Q: How would this change likely affect: Node expansions? Runtime?

A: Increasing $h(D)$ from 3 to 7 would make $D$ look less promising because its $f(n) = g(n) + h(n)$ value would increase. As a result, A* would be less likely to expand $D$ early. Since this new heuristic overestimates the true cost from $D$ to $G$, it is no longer admissible, meaning A* may lose its guarantee of finding the optimal path. In general, if changing a heuristic value makes the heuristic no longer admissible, A* may become less reliable because it can be misled away from the optimal path, even if the runtime or node expansions may not necessarily increase.

#### (d) In one or two sentences, why can A* still approach exponential behavior in large graphs, even with an admissible heuristic?
A* can still approach exponential behavior because the number of possible paths grows rapidly with the branching factor and search depth. Even with an admissible heuristic, if the heuristic is weak or not very informative, A* may still need to expand many or nearly all relevant nodes before the optimal has been found.
