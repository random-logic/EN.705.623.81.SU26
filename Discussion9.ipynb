# 1

One example of an AI system that makes decisions over time is a navigation app, such as Google Maps or Waze, that helps a driver reach a destination.

## Agent

The agent is the navigation system. Its job is to choose directions or route recommendations for the driver.

## Environment

The environment includes the road network, traffic conditions, weather, accidents, construction zones, speed limits, and the driver’s current location. The environment can change over time as traffic gets heavier, roads close, or new delays appear.

## Possible Actions

The agent’s possible actions include:

- Recommend turning left, right, or continuing straight
- Suggest taking a highway or local road
- Recalculate the route
- Warn the driver about traffic or hazards
- Suggest an alternative route

## States or Information the Agent Should Pay Attention To

The agent should consider information such as:

- Current location of the driver
- Destination
- Current traffic speeds
- Road closures or accidents
- Estimated travel time for different routes
- Distance remaining
- Time of day
- Weather conditions
- Driver preferences, such as avoiding tolls or highways

## Rewards or Goals

The main goal is to get the driver to the destination efficiently and safely. Rewards could be based on:

- Shorter travel time
- Shorter distance
- Avoiding traffic delays
- Avoiding dangerous roads or accidents
- Following user preferences
- Reducing fuel use

A positive reward might be given for reaching the destination quickly and safely. Negative rewards might be given for delays, unsafe routes, missed turns, or sending the driver through heavy traffic.

## Why This Is a Sequential Decision-Making Problem

This is a sequential decision-making problem because the navigation app must make many decisions over time, not just one prediction. Each recommendation affects what happens next. For example, choosing one road may lead to faster travel now, but it might also cause the driver to encounter traffic later. The app must keep updating its decisions as the driver moves and as traffic conditions change.

A one-step prediction problem might only estimate the travel time of a route at one moment. In contrast, a navigation system must repeatedly choose actions, observe the results, and adjust future recommendations.

## Role of a Policy

A policy would help the navigation agent decide what action to take in each situation. For example, the policy might say:

- If there is heavy traffic ahead, suggest an alternate route.
- If the driver misses a turn, recalculate the route.
- If the highway is faster and the user allows highways, recommend the highway.
- If a road is closed, avoid that road.

The policy maps the current state of the environment to an action. Over time, a good policy helps the agent make decisions that lead to the best overall outcome.

## Immediate vs. Future Rewards

Future rewards are very important in navigation. A route that looks good immediately may not be best overall. For example, a shortcut might save two minutes now but lead to a major traffic jam later. On the other hand, taking a slightly longer road now might avoid congestion and result in a faster total trip.

Because of this, the navigation app should consider both immediate rewards, such as saving time right now, and future rewards, such as arriving at the destination sooner and more safely overall.

# 2
## States

A state should include the important information the navigation app needs to make a decision at a given time. In this example, a state could include:

- The driver’s current location
- The destination
- Current traffic conditions on nearby roads
- Road closures or accidents
- Estimated travel times on possible routes
- Weather conditions
- Time of day
- Driver preferences, such as avoiding tolls or highways

A state represents the current situation the navigation app observes before recommending the next action.

## Actions

The navigation app’s possible actions are the choices it can recommend to the driver. These may include:

- Continue straight
- Turn left
- Turn right
- Take a highway
- Take a local road
- Reroute around traffic
- Avoid a road closure
- Recommend stopping or slowing due to hazards

Each action changes the driver’s future location and affects the route options available later.

## Transition Behavior

The transition behavior describes how the environment changes after the agent chooses an action. In a navigation app, transitions are uncertain because traffic and road conditions can change.

For example:

- If the app recommends taking the highway, the driver may move closer to the destination quickly if traffic is light.
- If traffic suddenly increases, the same action may lead to a delay.
- If the driver misses a turn, the next state will be different from what the app expected.
- If an accident occurs ahead, the app may need to reroute.

So, the transition function gives the probability of moving from one state to another after taking a certain action.

## Reward Structure

The reward tells the agent how good or bad an action is. For a navigation app, rewards could be designed around the overall goal of reaching the destination efficiently and safely.

Possible rewards include:

- Positive reward for reaching the destination
- Negative reward for each minute spent traveling
- Negative reward for traffic delays
- Negative reward for unsafe or risky routes
- Negative reward for missing turns or taking inefficient roads
- Positive reward for following user preferences, such as avoiding tolls

The agent should try to maximize total reward over the whole trip, not just the reward from the next turn.

## Markov Property

The Markov property means that the current state contains all the information needed to predict what happens next, without needing the full history of past states.

In this example, the state definition mostly satisfies the Markov property if it includes the driver’s current location, destination, traffic conditions, road closures, weather, time of day, and relevant user preferences. If all of this information is included, then the navigation app does not need to know the exact route history to make a good next decision. However, the Markov property may not be perfect because some hidden information, such as future accidents, driver habits, or unreported road conditions, may affect what happens next.

## State Value

The state value represents how good it is for the driver to be in a certain state, assuming the navigation app follows a particular policy from that point forward.

For example, a state where the driver is close to the destination, traffic is light, and roads are clear would have a high value. A state where the driver is far away, stuck in traffic, and surrounded by road closures would have a lower value.

## Action Value

The action value represents how good it is to take a specific action in a specific state and then continue following a policy afterward.

For example, if the driver is approaching an intersection, the action value of turning left might be high if it avoids traffic and leads to a faster route. The action value of continuing straight might be lower if it leads into congestion. Action values help the navigation app compare possible choices from the same state.

## Bellman Equation and Long-Term Decision Quality

The Bellman equation represents long-term decision quality by saying that the value of a state depends on both the immediate reward and the expected value of future states. In the navigation example, this means a route choice should not be judged only by whether it saves time right now, but also by where it leaves the driver afterward. A turn that looks slower at the moment might have a higher long-term value if it avoids future traffic or road closures. The Bellman equation helps the agent break a large route-planning problem into smaller decision steps while still considering the total expected outcome. This formal MDP view goes beyond the intuitive Module 9A description by clearly defining states, actions, transition probabilities, rewards, and value functions. It gives a mathematical structure for evaluating and improving navigation decisions over time.

# 3

We are given the following MDP:

- States: $S1, S2, S3, Goal$
- Terminal state: $Goal$ with value $0$
- Discount factor: $\gamma = 0.9$
- Initial values:  
  $V_0(S1) = 0$  
  $V_0(S2) = 0$  
  $V_0(S3) = 0$  
  $V_0(Goal) = 0$

The value iteration update is:

$V_{k+1}(s) = \max_a [R(s,a) + \gamma V_k(s')]$

where $s'$ is the next state reached after taking action $a$.



## Iteration 1

Using the initial values $V_0 = 0$:

### State $S1$

Action $a$:

$2 + 0.9V_0(S2) = 2 + 0.9(0) = 2$

Action $b$:

$1 + 0.9V_0(S3) = 1 + 0.9(0) = 1$

So:

$V_1(S1) = \max(2, 1) = 2$



### State $S2$

Action $a$:

$5 + 0.9V_0(Goal) = 5 + 0.9(0) = 5$

Action $b$:

$0 + 0.9V_0(S3) = 0 + 0.9(0) = 0$

So:

$V_1(S2) = \max(5, 0) = 5$



### State $S3$

Action $a$:

$3 + 0.9V_0(Goal) = 3 + 0.9(0) = 3$

Action $b$:

$1 + 0.9V_0(S2) = 1 + 0.9(0) = 1$

So:

$V_1(S3) = \max(3, 1) = 3$



## Values After Iteration 1

$V_1(S1) = 2$

$V_1(S2) = 5$

$V_1(S3) = 3$

$V_1(Goal) = 0$



## Iteration 2

Now use the values from iteration 1.

### State $S1$

Action $a$:

$2 + 0.9V_1(S2) = 2 + 0.9(5) = 2 + 4.5 = 6.5$

Action $b$:

$1 + 0.9V_1(S3) = 1 + 0.9(3) = 1 + 2.7 = 3.7$

So:

$V_2(S1) = \max(6.5, 3.7) = 6.5$



### State $S2$

Action $a$:

$5 + 0.9V_1(Goal) = 5 + 0.9(0) = 5$

Action $b$:

$0 + 0.9V_1(S3) = 0 + 0.9(3) = 2.7$

So:

$V_2(S2) = \max(5, 2.7) = 5$



### State $S3$

Action $a$:

$3 + 0.9V_1(Goal) = 3 + 0.9(0) = 3$

Action $b$:

$1 + 0.9V_1(S2) = 1 + 0.9(5) = 1 + 4.5 = 5.5$

So:

$V_2(S3) = \max(3, 5.5) = 5.5$



## Values After Iteration 2

$V_2(S1) = 6.5$

$V_2(S2) = 5$

$V_2(S3) = 5.5$

$V_2(Goal) = 0$



## Policy Extraction

Using the updated values after the second iteration, we choose the action in each state that gives the highest expected return.

### Best Action from $S1$

Action $a$:

$2 + 0.9V_2(S2) = 2 + 0.9(5) = 6.5$

Action $b$:

$1 + 0.9V_2(S3) = 1 + 0.9(5.5) = 5.95$

Best action:

$\pi(S1) = a$



### Best Action from $S2$

Action $a$:

$5 + 0.9V_2(Goal) = 5$

Action $b$:

$0 + 0.9V_2(S3) = 0 + 0.9(5.5) = 4.95$

Best action:

$\pi(S2) = a$



### Best Action from $S3$

Action $a$:

$3 + 0.9V_2(Goal) = 3$

Action $b$:

$1 + 0.9V_2(S2) = 1 + 0.9(5) = 5.5$

Best action:

$\pi(S3) = b$



## Final Policy

$\pi(S1) = a$

$\pi(S2) = a$

$\pi(S3) = b$

So the best actions are:

- From $S1$, take action $a$
- From $S2$, take action $a$
- From $S3$, take action $b$



## Explanation

This activity illustrates Bellman optimality because each state value is updated by choosing the action that gives the largest combination of immediate reward and discounted future value. For example, $S3$ eventually prefers action $b$ because going to $S2$ produces a better long-term return than going directly to the goal. Value iteration supports model-based planning because it uses the known transition and reward structure of the MDP to estimate the best long-term value of each state. Instead of learning only from trial and error, the agent can plan ahead by repeatedly applying Bellman updates. After enough iterations, the values move closer to the optimal values, and the best policy can be extracted by choosing the action with the highest expected return in each state.

# 4

We have:

- States: $s_0, s_1$
- Actions: $a_0, a_1$
- Discount factor: $\gamma = 0.9$
- Learning rate: $\alpha = 1.0$
- Initial Q-values: all zeros

The Q-learning update rule is:

$Q(s,a) \leftarrow Q(s,a) + \alpha [r + \gamma \max_{a'} Q(s',a') - Q(s,a)]$

Since $\alpha = 1.0$, the update simplifies to:

$Q(s,a) \leftarrow r + \gamma \max_{a'} Q(s',a')$

|------:|------:|------:|
| $s_0$ |   $0$ |   $0$ |
| $s_1$ |   $0$ |   $0$ |



## (b) and (c) Q-Learning Updates

## Update 1: $(s_0, a_0) \rightarrow s_0$, reward $r = 1$

Current value:

$Q(s_0,a_0) = 0$

Next-state maximum:

$\max_{a'} Q(s_0,a') = \max(Q(s_0,a_0), Q(s_0,a_1)) = \max(0,0) = 0$

Update:

$Q(s_0,a_0) \leftarrow 1 + 0.9(0) = 1$

### Q-Table After Update 1

|-------|------:|------:|
| $s_0$ |   $1$ |   $0$ |
| $s_1$ |   $0$ |   $0$ |



## Update 2: $(s_0, a_1) \rightarrow s_1$, reward $r = 0$

Current value:

$Q(s_0,a_1) = 0$

Next-state maximum:

$\max_{a'} Q(s_1,a') = \max(Q(s_1,a_0), Q(s_1,a_1)) = \max(0,0) = 0$

Update:

$Q(s_0,a_1) \leftarrow 0 + 0.9(0) = 0$

## Update 3: $(s_1, a_0) \rightarrow s_0$, reward $r = 4$

Current value:

$Q(s_1,a_0) = 0$

Next-state maximum:

$\max_{a'} Q(s_0,a') = \max(Q(s_0,a_0), Q(s_0,a_1)) = \max(1,0) = 1$

$Q(s_1,a_1) = 0$

$Q(s_1,a_1) \leftarrow 2 + 0.9(4.9)$

## (e) Short Responses

### Which state-action pair ended with the highest Q-value, and why?

The state-action pair $(s_1, a_1)$ ended with the highest Q-value:

$Q(s_1,a_1) = 6.41$

This happened because action $a_1$ from $s_1$ gives an immediate reward of $2$ and keeps the agent in $s_1$, where the best known future action has value $4.9$. Therefore, its Q-value includes both the immediate reward and a strong discounted future reward.

### How does the update rule incorporate possible future reward?

The update rule incorporates possible future reward through the term:

$\gamma \max_{a'} Q(s',a')$

This term looks at the best estimated Q-value available in the next state $s'$, then discounts it by $\gamma$. In this case, $\gamma = 0.9$, so future rewards are considered important but slightly less valuable than immediate rewards.

### Why is Q-learning considered model-free?

Q-learning is considered model-free because the agent does not need to know the full transition model or reward model of the environment in advance. It does not need probabilities for moving between states. Instead, it updates Q-values directly from experience by observing the current state, action, reward, and next state.
