## Background on Algorithm Analysis (50 points total)

- Constant Time (O(1)): Regardless of the input size, the algorithm takes a constant amount of time to complete.
    - Accessing an element in an array by index.
    - Inserting or deleting an element at the beginning of a linked list.


- Linear Time (O(n)): The runtime grows linearly with the size of the input.
    - Iterating through a list or array once.
    - Searching for an element in an unsorted list by traversing it.


- Logarithmic Time (O(log n)): As the input size grows, the runtime increases logarithmically.
    - Binary search in a sorted array.
    - Operations in a balanced binary search tree (BST).
    

- Quadratic Time (O(n^2)): The runtime grows quadratically with the input size.
    - Nested loops where each loop iterates through the input.
    - Some sorting algorithms like Bubble Sort or Selection Sort.


- Exponential Time (O(2^n)): The runtime doubles with each addition to the input size.
    - Algorithms involving recursive solutions that make repeated calls.
    - Solving the Traveling Salesman Problem using brute force.
    

- Factorial Time (O(n!)): The runtime grows factorially with the input size.
    - Permutation problems that explore all possible combinations, like the brute force solution for the Traveling Salesman Problem with no optimizations.

## Bubble Sort (Example)

Use the below analysis and explanation as a template for answering the following questions on algorithm analysis. You are to analyze each algorithm line by line, calculate the runtime complexity, and provide an explanation in markdown.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):  # O(n)
        for j in range(0, n-i-1):  # O(n)
            if arr[j] > arr[j+1]:  # O(1)
                arr[j], arr[j+1] = arr[j+1], arr[j]  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n^2)
# Big Omega notation: Ω(n)
# Big Theta notation: Θ(n^2)
```

Explanation:

- The outer loop iterates 'n' times, where 'n' is the length of the array.
- The inner loop also iterates 'n' times in the worst case.
- The comparisons and swapping operations inside the inner loop are constant time ('O(1)').

This results in a total runtime complexity of O(n^2) in the worst case, Ω(n) in the best case (already sorted), and Θ(n^2) in the average and worst cases, as both loops iterate over the input array.

## 1A Merge Sort (10 points)

```python
def merge_sort(arr):  # 2 * T(n / 2) + O(n)
    if len(arr) > 1:  # O(1)
        mid = len(arr) // 2  # O(1)
        left_half = arr[:mid]  # O(n)
        right_half = arr[mid:]  # O(n)

        merge_sort(left_half)  # T(n / 2)
        merge_sort(right_half)  # T(n / 2)

        i = j = k = 0  # O(1)

        while i < len(left_half) and j < len(right_half):  # O(n)
            if left_half[i] < right_half[j]:  # O(1)
                arr[k] = left_half[i]  # O(1)
                i += 1  # O(1)
            else:
                arr[k] = right_half[j]  # O(1)
                j += 1  # O(1)
            k += 1  # O(1)

        while i < len(left_half):  # O(n)
            arr[k] = left_half[i]  # O(1)
            i += 1  # O(1)
            k += 1  # O(1)

        while j < len(right_half):  # O(n)
            arr[k] = right_half[j]  # O(1)
            j += 1  # O(1)
            k += 1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n log n)
# Big Omega notation: Ω(n log n)
# Big Theta notation: Θ(n log n)
```

Explanation:

Merge sort works by repeatedly dividing the input array into two halves until each subarray has only one element. An array of one element is already sorted. After dividing, the algorithm merges the sorted halves back together. During the merge step, each element from the two halves is compared and copied back into the original array. This merge step takes linear time, or $O(n)$, for an array of size $n$.

At each level of recursion, the total amount of work done by merging is $O(n)$ because all elements are processed once on that level. The array is divided in half each time, so there are $log n$ levels of recursion.

Therefore, the total time complexity is:
$$O(n) \times O(\log n) = O(n \log n)$$

Since merge sort always divides the array into halves and always performs the merge step regardless of whether the array is already sorted or not, its best-case, average-case, and worst-case running times are all:

$$\Theta(n \log n)$$

So the Big-O, Big-Omega, and Big-Theta complexities are:
$$O(n \log n), \quad \Omega(n \log n), \quad \Theta(n \log n)$$

## 1B Binary Search (10 points)

```python
def binary_search(arr, target):
    low = 0  # O(1)
    high = len(arr) - 1  # O(1)

    while low <= high:  # O(log n)
        mid = (low + high) // 2  # O(1)
        mid_val = arr[mid]  # O(1)

        if mid_val == target:  # O(1)
            return mid  # O(1)
        elif mid_val < target:  # O(1)
            low = mid + 1  # O(1)
        else:
            high = mid - 1  # O(1)

    return -1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(log n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(log n)
```

Explanation:

Binary search works on a sorted list by repeatedly checking the middle element of the current search range.

If the middle element is the target, the search ends immediately. In the best case, the target is the middle element the first time, so the best-case time complexity is:
$$\Omega(1)$$

If the target is not the middle element, the algorithm decides whether to continue searching in the left half or the right half. It can do this because the list is sorted. Each step cuts the remaining search space in half. For example, the number of possible elements goes from:
$$n,\frac{n}{2},\frac{n}{4},\frac{n}{8}, \dots, 1$$

The number of times $n$ can be divided by 2 before reaching 1 is:
$$\log_2 n$$

Therefore, in the worst case, binary search takes:
$$O(\log n)$$

On average, the algorithm still eliminates half of the remaining elements at each step, so the big theta complexity is:
$$\Theta(\log n)$$

Therefore, all the complexities are:
$$O(\log n), \quad \Omega(1), \quad \Theta(\log n)$$

## 1C Linear Search (10 points)

```python
def linear_search(arr, target):
    for i in range(len(arr)):  # O(n)
        if arr[i] == target:  # O(1)
            return i  # O(1)
    return -1  # O(1)

# Total Runtime Complexity:
# Big O notation: O(n)
# Big Omega notation: Ω(1)
# Big Theta notation: Θ(n)
```

Explanation:

Linear search checks each element in the array one by one until it either finds the target or reaches the end of the array.

In the best case, the target is the first element in the array. The algorithm only needs one comparison, so the best-case time complexity is:
$$\Omega(1)$$

In the worst case, the target is either the last element or is not in the array at all. In that situation, the algorithm must check every element, so the worst-case time complexity is:
$$O(n)$$

In the average case, the target is expected to be somewhere around the middle of the array. The algorithm checks about $n/2$ elements on average. So the average time is:
$$\Theta(n)$$

Therefore, all the complexities are:
$$O(n), \quad \Omega(1), \quad \Theta(n)$$

## 1E Heap Sort (10 points)

```python
def heapify(arr, n, i):  # O(log n)
    largest = i  # O(1)
    left = 2 * i + 1  # O(1)
    right = 2 * i + 2  # O(1)

    if left < n and arr[left] > arr[largest]:  # O(1)
        largest = left  # O(1)

    if right < n and arr[right] > arr[largest]:  # O(1)
        largest = right  # O(1)

    if largest != i:  # O(1)
        arr[i], arr[largest] = arr[largest], arr[i]  # O(1)
        # It can recurse at most O(log n) times because the heap height is only log n.
        heapify(arr, n, largest)  # O(log n)


def heap_sort(arr):
    n = len(arr)  # O(1)

    # Building the heap
    for i in range(n // 2 - 1, -1, -1):  # O(n) total
        heapify(arr, n, i)  # O(log n) worst case per call

    # Repeatedly move the largest value in the heap to the end of the array.
    # After each swap, reduce the heap size and restore the max heap property.
    for i in range(n - 1, 0, -1):  # O(n log n)
        arr[i], arr[0] = arr[0], arr[i]  # O(1)
        heapify(arr, i, 0)  # O(log n)

# Total Runtime Complexity:
# Big O notation: O(n log n)
# Big Omega notation: Ω(n log n)
# Big Theta notation: Θ(n log n)
```

Explanation:

Heap sort works by using a binary heap, specifically a max heap. A max heap is a tree-like structure where every parent node is greater than or equal to its children. Because of this property, the largest value in the heap is always at the root. Even though we think of a heap as a tree, it is stored inside an array. For any element at index $i$, its left child is found at index $2i + 1$, and its right child is found at index $2i + 2$.

The heapify process is used to restore the max heap property at a given position. It compares a parent node with its left and right children. If one of the children is larger than the parent, the parent is swapped with the larger child. After the swap, the value that moved downward may still violate the heap property, so heapify continues downward until the subtree satisfies the max heap property again. A single heapify operation can take $O(\log n)$ time in the worst case because the value may have to move down the height of the heap. Since a binary heap has height $\log n$, heapify is $O(\log n)$.

Heap sort has two main phases.

In the first phase, it builds a max heap from the input array. This is done by calling heapify starting from the last non-leaf node and moving backward toward the root. Although each heapify call can take $O(\log n)$ in the worst case, building the heap takes $O(n)$ total time. This is because most nodes are near the bottom of the heap and can only move down a small number of levels. In a binary heap:

* $\frac{n}{2}$ nodes are leaves and require no heapify work.
* $\frac{n}{4}$ nodes are one level above the leaves and can move down at most 1 level.
* $\frac{n}{8}$ nodes are two levels above the leaves and can move down at most 2 levels.
* $\frac{n}{16}$ nodes are three levels above the leaves and can move down at most 3 levels.
* And so on...

So the total work to create the heap can be written as:

$$\frac{n}{2}(0) + \frac{n}{4}(1) + \frac{n}{8}(2) + \frac{n}{16}(3) + \cdots$$

This represents the number of nodes at each height multiplied by the maximum amount of work each node can do.

Factor out $n$:
$$n\left(\frac{0}{2} + \frac{1}{4} + \frac{2}{8} + \frac{3}{16} + \cdots\right)$$

The expression inside the parentheses is a constant:
$$\frac{0}{2} + \frac{1}{4} + \frac{2}{8} + \frac{3}{16} + \cdots = \sum_{i = 0}^\infty \frac{i}{2^{i + 1}} = 1$$

So the total work becomes:

$$O(n)$$

```python
def fibonacci_iterative(n):
    # Generates the first n Fibonacci numbers
    sequence = []  # O(1)

    a, b = 0, 1  # O(1)

    for _ in range(n):  # O(n)
        sequence.append(a)  # O(1)
        a, b = b, a + b  # O(1)

    return sequence  # O(1)


# Total Runtime Complexity:
# Big O notation: O(n)
# Big Omega notation: Ω(n)
# Big Theta notation: Θ(n)
```

Explanation:

The iterative Fibonacci algorithm generates the Fibonacci sequence by building it from the beginning. The Fibonacci sequence starts with $0$ and $1$. Each next number is found by adding the two previous numbers:
$$F(n) = F(n - 1) + F(n - 2)$$

For example:
$$0, 1, 1, 2, 3, 5, 8, 13, \dots$$

The iterative version keeps track of two values at a time: the current Fibonacci number and the next Fibonacci number. At each step, the current number is added to the sequence. Then the two tracking values are updated so the algorithm can move to the next Fibonacci number. The loop runs exactly $n$ times because it generates $n$ Fibonacci numbers. Each iteration does a constant amount of work, such as appending a number to the sequence and updating the two variables. Therefore, the total runtime is:
$$O(n)$$

Since the algorithm must generate $n$ numbers, it also has a lower bound of:
$$\Omega(n)$$

So the tight bound is:
$$\Theta(n)$$

The runtime complexities are:
$$\text{Big O: } O(n)$$
