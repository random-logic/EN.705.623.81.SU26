# 1

A familiar real-world system that can be modeled as a graph is a city public transit network. The nodes represent transit stops or stations. The edges represent direct routes between two stops. The edges are directed because some transit routes may only travel in one direction or may have different travel times depending on direction. The edges are weighted using estimated travel time in minutes between stops.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed weighted graph
G = nx.DiGraph()

# Add nodes representing transit stops/stations
stations = [
    "Central Station",
    "Main Street",
    "City Hall",
    "University",
    "Airport",
    "Museum",
    "Stadium"
]

G.add_nodes_from(stations)

# Add directed edges with weights representing travel time in minutes
routes = [
    ("Central Station", "Main Street", 5),
    ("Main Street", "City Hall", 4),
    ("City Hall", "University", 8),
    ("University", "Airport", 15),
    ("Central Station", "Museum", 7),
    ("Museum", "Stadium", 6),
    ("Stadium", "University", 10),
    ("Airport", "Central Station", 20),
    ("City Hall", "Central Station", 6),
    ("University", "City Hall", 9)
]

# Add edges programmatically
for start, end, time in routes:
    G.add_edge(start, end, weight=time)

# Print graph information
print("Nodes:")
print(G.nodes())

print("\nEdges with travel times:")
for u, v, data in G.edges(data=True):
    print(f"{u} -> {v}, travel time: {data['weight']} minutes")

# Optional: visualize the graph
pos = nx.spring_layout(G, seed=42)

edge_labels = nx.get_edge_attributes(G, "weight")

plt.figure(figsize=(10, 7))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=2500,
    font_size=9,
    arrows=True
)

nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_labels,
    font_size=8
)

plt.title("Directed Weighted Public Transit Network")
plt.show()
```

### A. Emergency Dispatch
Traversal algorithm: BFS

BFS, or Breadth-First Search, is the better choice because the AI system needs to find the nearest available emergency vehicle. Since all roads take approximately the same time to traverse, the graph can be treated as unweighted. BFS explores nodes level by level, meaning it checks all locations one road away, then two roads away, and so on. Therefore, the first available emergency vehicle found by BFS will be the closest in terms of number of road segments.

### B. Dependency Analysis
Traversal algorithm: DFS

DFS, or Depth-First Search, is the better choice for analyzing dependencies in a directed graph. DFS explores deeply along each dependency chain before backtracking, which makes it useful for detecting cycles in directed graphs. It can also help determine build or execution order through techniques like topological sorting, assuming the graph has no cycles. Because dependency analysis often requires following chains of relationships, DFS is conceptually well suited for this problem.

### C. Maze Exploration
Traversal algorithm: DFS

DFS is a good choice because the AI agent only needs to determine whether any path exists from the entrance to the exit. The agent does not need the shortest path, so BFS is not necessary. DFS explores one path as far as possible before backtracking, which is efficient for search and exploration when finding any valid route is sufficient. This makes DFS appropriate for maze exploration, especially when memory usage is a concern compared with BFS.

### A. Data Center Server Wiring
Graph algorithm: MST

This scenario is about connecting all server racks while minimizing the total cost of the wiring. The goal is not to find the cheapest route between two specific racks, but to create a connected network that includes every rack at the lowest total installation cost. That matches a Minimum Spanning Tree problem.

### B. Package Delivery Route
Graph algorithm: Shortest Path

This scenario is about finding the most efficient route from one specific starting point, the warehouse, to one specific destination, the customer location. Since the roads have different travel times and traffic conditions affect travel cost, the goal is to minimize the total travel time or cost along a path. That matches a Shortest Path problem.

### C. Smart Agriculture Sensor Network
Graph algorithm: MST

This scenario is about making sure all sensors are connected so they can transmit data while minimizing the total network cost. The goal is to connect every sensor in the network, not just find a route between two sensors. Because each wireless link has a cost and the objective is minimum total cost for full connectivity, this is a Minimum Spanning Tree problem.
